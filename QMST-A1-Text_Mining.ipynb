```python
#2a.
#!pip install wordcloud
from wordcloud import WordCloud
import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Read CSV file into Dataframe df
dataframe = pd.read_csv("game-of-thrones-deaths-data.csv", parse_dates=True, encoding="latin1")
pd.options.display.float_format = '{:.2f}'.format
dataframe = dataframe.dropna()    #removes empty data entries
```

```python
transcript = "Blood of my blood! You kept all your promises to me."

#delimiter = ''
l_words = transcript.split ()
l_words [:5]
```

```python
#count the words 
len(l_words) 
```

```python
import string 
```

```python
l_goodword = list() 
for word in l_words: 
    w_1 = word.lower() #make all letters into lower case 
    w_2 = w_1.strip(string.punctuation + string.whitespace)#remove all punctuation and
    l_goodword.append(w_2) 
 
l_goodword[:5] 
```

```python
d = dict() 
for word in l_goodword : 
    d[word] = d.get(word, 0) + 1 
d 
```

```python
#Alternative 
d = dict() 
for word in l_goodword : 
  if word in d : 
    d[word] += 1 
  else : 
    d[word] = 1  
d 
```

```python
import matplotlib.pyplot as plt 
```

```python
from wordcloud import WordCloud 
 
wordcloud = WordCloud(background_color='white', 
                      width=1500, 
                      height=1000 
                      ).generate_from_frequencies(d) 
 
plt.figure(figsize=(9,6)) 
plt.imshow(wordcloud) 
plt.axis('off') 
plt.show() 
```

```python
d_sorted = dict() 
 
sorted_value = sorted(d.values(), reverse = True) 
#If True, the sorted list is reversed (or sorted in descending order).  
#Defaults to False if not provided. 
 
for sortedKey in sorted_value: 
    for key, value in d.items(): 
        if value == sortedKey: 
            d_sorted[key] = value
 
d_sorted 
```

```python
# Initialize limit   
K = 10 
 
# Get first K items in dictionary   
d_select = dict(list(d_sorted.items())[: K])   
 
d_select 
```

```python
plt.bar(list(d_select.keys()), d_select.values(), color='r') 
plt.show() 
```
